# Biquad filter


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

# 从denoise.c文件中获取高通滤波器的系数
a_hp = np.array([1, -1.99599, 0.99600])
b_hp = np.array([1, -2, 1])

#   float weighting_b[2] = {-2.f, 1.f};
#   float weighting_a[2] = {-1.89f, .895f};
# a_hp = np.array([1, -1.89, 0.895])
# b_hp = np.array([1, -2, 1])

# 计算频率响应
w, h = signal.freqz(b_hp, a_hp, worN=8000)

# 绘制频率响应图
plt.figure(figsize=(10, 6))

# 绘制幅度响应
plt.subplot(2, 1, 1)
plt.plot(0.5*w/np.pi, 20*np.log10(np.abs(h)), 'b')
plt.title('Biquad Filter Frequency Response')
plt.ylabel('Amplitude [dB]')
plt.xlabel('Frequency [Normalized]')
plt.grid()

# 绘制相位响应
plt.subplot(2, 1, 2)
angles = np.unwrap(np.angle(h))
plt.plot(0.5*w/np.pi, angles, 'g')
plt.ylabel('Phase (radians)')
plt.xlabel('Frequency [Normalized]')
plt.grid()

plt.tight_layout()
plt.show()
```
